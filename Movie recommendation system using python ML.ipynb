```python
Specifications = ['title','genres','keywords','tagline','cast','director']
print(Specifications)
```

```python
for specs in Specifications:
    Movies[specs] = Movies[specs].fillna('')
```

```python
#combining the specs
```

```python
combined_specs = Movies['title']+' '+Movies['genres']+' '+Movies['keywords']+' '+Movies['tagline']+' '+Movies['cast']+' '+Movies['director']
```

```python
print(combined_specs)
```

```python
#will convert text data to numerical values
```

```python
vectorizer = TfidfVectorizer()
```

```python
specs_vectors = vectorizer.fit_transform(combined_specs)
```

```python
print(specs_vectors)
```

```python
#now to find the similarity score
```

```python
similarity = cosine_similarity(specs_vectors)
```

```python
print(similarity)
```

```python
print(similarity.size)
```

```python
print(similarity.shape)
```

```python
# basically it is taking one movie and comparing with all the others similarity score hence the same array
```

```python
#user input for movie name
```

```python
movie_name = input("Enter your favourite movie name :")
```

```python
#create a list containing all titles for the movies in the dataset
```

```python
Movie_titles = Movies['title'].tolist()
print(Movie_titles)
```

```python
#finding a close match for the movie
```

```python
find_close_match = difflib.get_close_matches(movie_name, Movie_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
#finding the index value of the movie so it is easy to locate on the table
```

```python
index_value = Movies[Movies.title == close_match]['index'].values[0]
print(index_value)
```

```python
#getting a list of similar movies based on similarity score
```

```python
similarity_score = list(enumerate(similarity[index_value]))
```

```python
print(similarity_score)
```

```python
#avengers has similarity score of 1, which means other movies the higher the similarity score the more similar it is to the avnegers movie
```

```python
#now we will sort the movies based on highest to lowest similarity score
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
#x[1] means the second value in the index
```

```python
#printing name of similar movies using the index
```

```python
print('Movies we suggest for you: \n')
i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = Movies[Movies.index==index]['title'].values[0]
    if (i<10):
                                                   print(i, '.', title_from_index)
                                                   i+=1
```

```python
#top 10 movies suggested by matching highest similarity score and the index value together
```

```python
movie_name = input("Enter your favourite movie name :")
Movie_titles = Movies['title'].tolist()
find_close_match = difflib.get_close_matches(movie_name, Movie_titles)
close_match = find_close_match[0]
index_value = Movies[Movies.title == close_match]['index'].values[0]
similarity_score = list(enumerate(similarity[index_value]))
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print('Movies we suggest for you: \n')
i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = Movies[Movies.index==index]['title'].values[0]
    if (i<10):
                                                   print(i, '.', title_from_index)
                                                   i+=1
```
